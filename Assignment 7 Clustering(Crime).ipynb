```python
#Import Libraries
import numpy as np
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
crime_hc.info()
```

```python
crime_hc2 = crime_hc.drop(['Unnamed: 0'], axis=1)
crime_hc2
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
# Normalized data frame (considering the numerical part of data)
crime_hc_norm = norm_func(crime_hc2.iloc[:,1:])
```

```python
# create dendrogram
plt.figure(figsize=(10, 7))  
dendrogram = sch.dendrogram(sch.linkage(crime_hc_norm, method='single'))
```

```python
# create dendrogram
plt.figure(figsize=(10, 7))  
dendrogram = sch.dendrogram(sch.linkage(crime_hc_norm, method='complete'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'single')
hc
```

```python
y=pd.DataFrame(hc.fit_predict(crime_hc_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
crime_hc_norm['clustersid']=hc.labels_
crime_hc_norm
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime_hc_norm['clustersid'],crime_hc_norm['UrbanPop'], c=hc.labels_)
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(crime_k_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Curve')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

# DBSCAN model

```python
#import Libraries
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing  import StandardScaler
```

```python
#importing the data set
crime_db = pd.read_csv("C:/Users/Siri/Desktop/Excelr/Assignments/7.Clustering/crime_data.csv")
```

```python
crime_db.info()
```

```python
crime_db.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
crime_db_norm=StandardScaler().fit_transform(crime_db)
crime_db_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_db_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
crime_db['clusters']=dbscan.labels_
crime_db
```

```python
crime_db.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime_db['clusters'],crime_db['UrbanPop'], c=dbscan.labels_)
```
